```python
# Q1.
# Write a function `add_numbers` that takes any number of arguments
# using *args and returns their sum.
# Call the function with:
#   add_numbers(1, 2, 3)
#   add_numbers(5, 10, 15, 20)
def add_numbers(*args):
  sum = 0
  for i in args:
    sum += i
  return sum
print(add_numbers(1, 2, 3))
print(add_numbers(5, 10, 15, 20))
```

```python
# Q2.
# Write a function `student_info` that accepts:
#   - name (required)
#   - age (required)
#   - any number of additional keyword arguments (**kwargs)
# Print the name, age, and all extra details in key : value format.

def student_info(**kwargs):
  for k,v in kwargs.items():
    print(f"{k}:{v}")

student_info(name='jayanth', age=22, school='GSEMS', salary='36,000')
```

```python
# Q4.
# Given:
#   x = 10
# Write a function that prints the value of x.
# Then write another function that changes the value of x to 20.
# Call both functions and observe the output.
# Explain the behavior using scope rules.

x=10

def printX():
  print(x)
def changeX():
  global x
  x = 20

printX()
changeX()
printX()
```

```python
# Q6.
# Given:
#   nums = [1, 2, 3, 4, 5, 6]
# Use map() to create a new list containing squares of all numbers.
# Do NOT use loops or list comprehension.

nums = [1, 2, 3, 4, 5, 6]
sq_nums = map(lambda x:x**2,nums)
list(sq_nums)
```

```python
# Q7.
# Given:
#   nums = [10, 15, 20, 25, 30]
# Use filter() to create a list of numbers divisible by 10.
# Do NOT use loops.

nums = [10, 15, 20, 25, 30]
divisible_10 = filter(lambda x: x%10 == 0, nums)
list(divisible_10)
```

```python
# Q8.
# Given:
#   nums = [1, 2, 3, 4, 5]
# Use reduce() to calculate the product of all elements.
# (Import reduce from functools.)
# Do NOT use loops.
from functools import reduce
nums = [1, 2, 3, 4, 5]
res = reduce(lambda x,y:x*y,nums,1)
res
```

```python
# Q9.
# Write a function `calculator` that accepts:
#   - an operation name as a string ("add", "sub", "mul", "div")
#   - any number of numeric arguments using *args
# Perform the operation on the arguments.
# Handle division by zero properly.

def calculator(operation, *args):
    if not args:
        return 0

    # Initialize result with the first argument
    result = args[0]

    for num in args[1:]:
        if operation == "add":
            result += num
        elif operation == "sub":
            result -= num
        elif operation == "mul":
            result *= num
        elif operation == "div":
            if num == 0:
                return "Error: Division by zero"
            result /= num
        else:
            return "Error: Invalid operation"

    return result

# Examples:
print(calculator("add", 10, 5, 2))
print(calculator("div", 20, 5, 2))
print(calculator("div", 10, 0))

```

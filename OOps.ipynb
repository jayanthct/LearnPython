```python
# Q2.
# Create a class `Counter`
# Use a CLASS VARIABLE to keep track of how many objects are created.
# Print the count after creating 3 objects.
# (Do NOT use global variables.)

class Counter:
  count = 0
  def __init__(self):
    Counter.count += 1
    print(f'created obj -> {Counter.count}')

  @classmethod
  def getCount(cls):
    return cls.count

c1 = Counter()
c2 = Counter()
c3 = Counter()

print(c1.getCount())
print(c2.getCount())
print(c3.getCount())
```

```python
# Q3.
# Create a class `BankAccount` with:
#   - account_holder
#   - balance (default = 0)
# Methods:
#   deposit(amount)
#   withdraw(amount) â†’ do not allow withdrawal if balance is insufficient
# Create an object and test all methods.

class BankAccount:
  def __init__(self, user, balance=0):
    self.account_holder = user
    self.balance = balance
    print(f'Account Created, Balance is {self.balance}...')

  def deposit(self, amount:int)->None:
    self.balance += amount
    print(f'Amount {amount} deposited...')
    print(f'Balance Updated {self.balance}...')

  def withdraw(self, amount:int)->None:
    if self.balance < amount:
      print('Amount is greater than Balance, Try Again')
      return;

    self.balance-=amount
    print(f'Amount {amount} withdrawed...')
    print(f'Balance Updated {self.balance}...')


b1 = BankAccount('jayanth', 14500)
b1.deposit(120)
b1.withdraw(14300)
b1.withdraw(500)
b1.withdraw(124)
```

```python
# Q4.
# Create a class `Rectangle` with:
#   length, breadth
# Create a method `area()`
# Create a subclass `Square` that inherits Rectangle
# Ensure Square only takes one side as input.

class Shape:
  def __init__(self, length=0, bredth=0):
    self.length = length
    self.bredth = bredth

  def area(self):
    return self.length * self.bredth

class Rectangle(Shape):
  def __init__(self, length=0, bredth=0):
    super().__init__(length, bredth)

class Square(Rectangle):
  def __init__(self, side):
    super().__init__(side, side)


r = Rectangle(5, 6)
sq = Square(5)

print(r.area())
print(sq.area())
```

```python
from typing import override
# Q5.
# Create a class `Employee` with:
#   name, salary
# Create a method `get_salary()`
# Create a subclass `Manager` that:
#   adds a bonus to salary
# Override `get_salary()` properly.

class Employee:
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary

  def get_salary(self):
    return self.salary

class Manager(Employee):
  def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = bonus

  @override
  def get_salary(self):
    return super().get_salary() + self.bonus

M1 = Manager('jayanth', 2500, 1500)
print(M1.get_salary())
```

```python
# Q6.
# Create a class `Person` with method `role()`
# Create subclasses `Student` and `Teacher`
# Each subclass should implement `role()` differently.
# Call `role()` using a common reference and observe polymorphism.

class Person:
    def role(self):
        pass

class Student(Person):
    def role(self):
        return "I am a Student: I focus on learning and completing assignments."

class Teacher(Person):
    def role(self):
        return "I am a Teacher: I focus on mentoring and grading student work."

def print_role(person_type):
    print(person_type.role())

# Creating instances
student_obj = Student()
teacher_obj = Teacher()

print_role(student_obj)
print_role(teacher_obj)
```

```python
# Q7.
# Create a class `Car`
# Make the attribute `speed` PRIVATE
# Provide methods to:
#   set_speed()
#   get_speed()
# Try accessing speed directly and observe what happens.

class Car:
  def __init__(self, speed=0):
    self.__speed = speed

  def set_speed(self, speed):
    self.__speed = speed
    print('speed set to', speed)

  def get_speed(self):
    return self.__speed

c= Car()
c.set_speed(5)
print(c.get_speed())
print(c.speed)
```

```python
# Q1.
# Create a dictionary `student` with keys:
#   "name", "age", "marks"
# Assign appropriate values.
# Print each key-value pair in the format:
# key : value
# (Do NOT print the dictionary directly.)

student = {
    'name' : 'jayanth',
    'age' : 24,
    'marks' : 144
}

for k,v in student.items():
  print(f"{k} : {v}")
```

```python
# Q2.
# Given:
#   data = {"a": 10, "b": 20, "c": 30}
# Increase each value by 5
# WITHOUT creating a new dictionary.
data = {"a": 10, "b": 20, "c": 30}
for k in data:
  data[k] += 5
print(data)
```

```python
# Q3.
# Given:
#   marks = {"math": 80, "science": 75, "english": 90}
# Find and print the subject with the highest marks
# WITHOUT using max().
marks = {"math": 80, "science": 75, "english": 90}
max = -1
subject = None
for k in marks:
  if marks[k] > max:
    max = marks[k]
    subject = k
print(f"{subject} : {max}")
```

```python
# Q4.
# Given:
#   info = {"name": "Alex", "city": "Delhi"}
# Add a new key "country" with value "India"
# Update "city" to "Mumbai"
# Remove the key "name"
# Print the final dictionary.
info = {"name": "Alex", "city": "Delhi"}
info['country'] = 'India'
info['city'] = 'Mumbai'
del info['name']
print(info)
```

```python
# Q5.
# Given:
#   d = {"x": 1, "y": 2}
# Access the value of key "z"
# Do NOT use try-except.
# Ensure the program does NOT crash.
d = {"x": 1, "y": 2}
print(d.get('z'))
```

```python
# Q6.
# Given:
#   scores = {"A": 90, "B": 85, "C": 88}
# Create two lists:
#   keys_list  → contains all keys
#   values_list → contains all values
# Do NOT use list() directly on the dictionary.
scores = {"A": 90, "B": 85, "C": 88}
keys = list(scores.keys())
vals = list(scores.values())
print(keys, vals)
```

```python
# Q7.
# Given:
#   data = {"a": 1, "b": 2, "c": 3}
# Create a new dictionary where keys and values are swapped.
# (Assume all values are unique.)
data = {"a": 1, "b": 2, "c": 3}
new_dict = {v: k for k, v in data.items()}
print(new_dict)
```

```python
# Q8.
# Given:
#   inventory = {"apple": 10, "banana": 5, "orange": 8}
# If "banana" exists:
#   decrease its quantity by 2
# Else:
#   add "banana" with quantity 2
# Print the updated dictionary.
inventory = {"apple": 10, "banana": 5, "orange": 8}
if('banana' in inventory):
  inventory['banana'] -= 2
else:
  inventory['banana'] = 2
print(inventory)
```

```python
# Q9.
# Write a program to count the frequency of each character
# in a given string using a dictionary.
# Example:
#   input: "banana"
#   output: {'b': 1, 'a': 3, 'n': 2}

input = "banana"
output = {}
for char in input:
  count = output.get(char, 0)
  output[char] = count + 1
print(output)
```

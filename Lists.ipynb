```python
# Q3.
# Given:
#   nums = [1, 2, 3, 4, 5, 6]
# Create a new list `even_nums` containing only even numbers
# WITHOUT using list comprehension.
nums = [1, 2, 3, 4, 5, 6]
even_nums =[]
for i in nums:
  if i%2==0:
    even_nums.append(i)
print(even_nums)

even = [i for i in nums if i%2==0]
print(even)
```

```python
# Q4.
# Given:
#   nums = [5, 10, 15, 20, 25]
# Reverse the list
# WITHOUT using reverse() or slicing ([::-1]).
nums = [5, 10, 15, 20, 25]
print(nums[::-1])
nums.reverse()
print(nums)
nums = [5, 10, 15, 20, 25]

i,j = 0, len(nums)-1
while i<j:
  nums[i], nums[j] = nums[j], nums[i]
  i+=1
  j-=1
print(nums)
```

```python
# Q5.
# Given:
#   a = [1, 2, 3]
#   b = a
# Append 4 to `a`
# Print both a and b
# Explain in comments why both changed or did not change.
import copy
a = [1, 2, 3]
b = copy.deepcopy(a)
a.append(4)
print(a)
print(b)
```

```python
# Q6.
# Given:
#   nums = [1, 2, 3, 4, 5]
# Remove all odd numbers from the list
# WITHOUT creating a new list.
# (Be careful while modifying a list during iteration.)
nums = [1,2,3,4,5]
for i in nums:
  if(i&1):
    nums.remove(i)
nums
```
